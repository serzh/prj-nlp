Scrape wikipedia now

```python
response = requests.get(wiki_page)
soup = BeautifulSoup(response.text, 'html5lib')
sd = soup.find("div", {"class": "mw-parser-output"})
text =  " ".join([i.text for i in sd.findAll("p")])
doc = nlp(text)
print("\n".join([sent.text for sent in doc.sents]))
```

The following are set of rules that extract movies from text:

```python
result = []
def is_year(year):
    #checks whether a candidate is a year
    if '–' in year:
        #e.g. "2007-10"
        year = year.split("-")[0]
    if len(year) == 4:
        try:
            if int(year) > 1900:
                return True
        except:
            pass

        return True

def match_movie_by_sub_year(ent):
    """
    this rule assumes that the NER is a movie if there is year in subtree
    e.g.  Notable films during this period include Avalon (1990), Paradise (1991)
    """

    movie = ent
    if ent.label_ != "DATE":
        for i in ent.subtree:
            if i.pos_ == "NUM":
                if is_year(i.text) and i.text != ent.text:
                    result.append((movie.text, i.text))
                    break

def date_contains_all_or_both_words(ent):
    """
    this rule checks if parent of year is ("all", "both"), than matches the movies
    e.g.  Radio Flyer, Forever Young (both 1992), The Adventures of Huck Finn and The Good Son (both 1993).
    Sin City, Green Street, Everything Is Illuminated (all 2005
    """
    if ent.label_ == "DATE":
        if ent[0].ancestors.__next__().text in ("all", "both"):
            res = [(b.text, ent.text) for b in [i for i in [i for i in ent[0].ancestors][2].children if i.dep_ == 'appos']]

def get_movies_from_entities(doc):
    for ent in doc.ents:
        match_movie_by_sub_year(ent)
        date_contains_all_or_both_words(ent)

def get_by_year(doc):
    #e.g. In 1999, Wood played a suburban white teenager who affects hip-hop lingo in James Toback's Black and White, and a junior hitman in Chain of Fools.
    for sent in doc.split("."):
        #spacy sometimes wrongly finds sentences
        year = "########"
        words = sent.split()
        if words[0] == "In":
            year = words[1].strip(",")
        elif " ".join(words[:2]) in ("Also in", "As of"):
            year = words[2].strip(",")
        if not is_year(year):
            continue
        title = []
        for word in words[::-1]:
            if word in ("in", "film", "season of") or word.endswith("'s"):
                break
            title.insert(0, word)
        if title:
#             print("======"," ".join(title))
            result.append((" ".join(title), year))
#         doc = nlp(sent)
#         print([ent for ent in doc.ents])
            
```
