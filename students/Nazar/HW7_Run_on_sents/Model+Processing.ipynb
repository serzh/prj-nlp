```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir('data') if isfile(join('data', f))]
```

```python
import gzip
ngrams={}
for file in onlyfiles:
    print(file)
    with gzip.open('data/'+file,'r') as f:
        for line in f:
            try:
                ngram, value = line.decode().split('\t')
                ngram_l=ngram.lower()
                if ngram_l in ngrams:
                    ngrams[ngram_l]=ngrams[ngram_l]+int(value[:-1])
                else:
                    ngrams[ngram_l]=int(value[:-1])
            except:
                print(line.decode())
```

```python
def get_bigram_count(word_1, word2):
    try:
        return ngrams[word_1+" "+ word2]
    except:
        return 0
```

```python
def get_bigram_end(word):
    try:
        return ngrams[word_1+" ."]
    except:
        return 0
```

```python
def get_bigram_end1(word):
    try:
        return ngrams[word_1+" ?"]
    except:
        return 0
```

```python
def get_bigram_end2(word):
    try:
        return ngrams[word_1+" !"]
    except:
        return 0
```

```python
import random
import re

def replace_twitter_name_with_normal_one(text):
    names_list=['He','he','she', 'She', 'It', 'we', 'We', 'they', 'i','I', 'you', 'u', 'You', 'Tom', 'Bill' , 'Mazar', 'Mike', 'nazar']
    index = random.randint(0, len(names_list)-1)
    new_name=names_list[index]
    return re.sub(r'@[a-z0-9A-Z]*', new_name, text.strip())

replace_twitter_name_with_normal_one('@Kenichan I dived many times for the ball. Managed to save 50%  The rest go out of bounds')
```

```python
import numpy as np
import tqdm

def transform_tweets_to_train_format(texts):
    train_data={'words':[],'labels':[]}
    for text_index in tqdm.tqdm(range(len(texts))):
        words_list=[]
        labels_list=[]
        sentenses = tokenize.sent_tokenize(texts[text_index])
        for sentens_index, sentense in enumerate(sentenses): 
            if len(sentense)<2:
                continue
            words = tokenize.word_tokenize(sentense if sentense[-1] not in "?!;.¸‚" else sentense[:-1])
            values = [False]*len(words)
            try:
                values[-1]=len(sentenses)>(sentens_index+1)
            except:
                print('Data:' ,texts[text_index])
                raise ValueError('A very specific bad thing happened.')
            words_list=words_list+words
            labels_list=labels_list+values
            
        if len(words_list)>0:
            train_data['words'].append(words_list)
            train_data['labels'].append(labels_list)

    return train_data
```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer(lang='uk')
word_info = morph.parse('серце')[0]
print('{0}: {1}, {2}, {3}({4}, )'.format('',word_info.tag.POS,word_info.normal_form,' ',' '))
```

```python
def merge_two_dicts(x, y):
    z = x.copy()   # start with x's keys and values
    z.update(y)    # modifies z with y's keys and values & returns None
    return z
```

```python
def get_word_features(sample, name=''):
    word=sample
    try:
        word_info=morph.parse(word)[0]
        postag = word_info.tag.POS
        normal_form=word_info.normal_form
    except:
        postag = "NoTAG"
        normal_form=word
    
    if postag==None:
        postag = "NoTAG"
    
    features = {
        name+'bias': 1.0,
        name+'word.normal_form()': normal_form,
        name+'word[-3:]': word[-3:],
        name+'word[-2:]': word[-2:],
        name+'word.isupper()': word.isupper(),
        name+'word.istitle()': word.istitle(),
        name+'word.isdigit()': word.isdigit(),
        name+'postag': postag,
        name+'postag[:2]': postag[:2]
    }
    return features

def get_features(word, lefts, rights):
    features =  get_word_features(word)
    for i, left in enumerate(lefts):
        features = merge_two_dicts(features, get_word_features(left,'left_{0}_'.format(i)))
    for i, right in enumerate(rights):
        features = merge_two_dicts(features, get_word_features(right,'right_{0}_'.format(i)))
    if len(rights)>0:
        bigram_count=get_bigram_count(word, rights[0] )
    else:
        bigram_count=0 
    
    bigram_end1=get_bigram_end1(word)
    bigram_end2=get_bigram_end2(word)
    bigram_end=get_bigram_end(word)
    features['bigram_stat']=round(bigram_count/(bigram_end1+bigram_end+1)/1000000000,2)

    return features
```

```python
def get_all_features(tokens):
    max_left=1
    max_right=1
    samples_count=len(tokens)
    featurs = []
    for index, sample in enumerate(tokens):
        if (index-max_left) >= 0:
            lefts =  tokens[(index-max_left):index]
        else:
            lefts=[]
        
        if (index+max_right+1) < samples_count:
            rights =  tokens[(index+1):(index+max_right+1)]
        else:
            rights=[]
            
        featurs.append(get_features(sample, lefts, rights))
    return  featurs

train_features = get_all_features(train['words'][0])
```
